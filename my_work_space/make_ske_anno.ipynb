```python



# 将所有pkl合成一个大的
pkl_dir = '/mmaction2/filling/val7_ske_annos'
dst_pkl_filename = '/mmaction2/filling/val_label0-3.pkl'

pkl_file_list = os.listdir(pkl_dir)


pkl_list = []
with open(dst_pkl_filename, 'wb') as f:
    for pkl_filename in pkl_file_list:
        # get label 

        label = int(pkl_filename.split('.')[0].split('_')[-1])
        if label == 4 or label == 5 or label == 6:
            continue
            # label = 1 # 1 代表其他
        with open(os.path.join(pkl_dir, pkl_filename), 'rb') as f_pkl:
            b = pickle.load(f_pkl)
            b['label'] = label
            pkl_list.append(b)
    pickle.dump(pkl_list, f)

pkl_list
```

```python
# 给文件夹下文件名的点号去掉。（不包括后缀）

p = '/mmaction2/filling/train3_ske_annos'

filenames = os.listdir(p)
for filename in filenames:
    pre = filename.rsplit('.', 1)[0]
    suf = filename.rsplit('.', 1)[-1]
    new_filename = pre.replace('.', '') + '.' + suf
    print(f'old filename : {filename}; new filename: {new_filename}')
    os.rename(os.path.join(p, filename), os.path.join(p, new_filename))
```
